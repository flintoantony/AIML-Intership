10. Print elements at even index positions

```python
numbers = [10, 20, 30, 40, 50]
for i in range(len(numbers)):
    if i % 2 == 0:
        print(numbers[i])
```

16. Print only even numbers

```python
numbers = [1, 2, 3, 4, 5, 6]
for num in numbers:
    if num % 2 == 0:
        print(num)
```

17. Print only odd numbers

```python
numbers = [1, 2, 3, 4, 5, 6]
for num in numbers:
    if num % 2 != 0:
        print(num)
```
